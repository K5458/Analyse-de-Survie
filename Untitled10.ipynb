```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from lifelines import KaplanMeierFitter, CoxPHFitter
from lifelines.statistics import logrank_test
from docx import Document

# Charger les données
data = pd.read_csv("C:/Users/HP/Documents/Test/Breast_Cancer.csv")  # Remplacez par le chemin de votre fichier

# Aperçu des données
print(data.head())
```

```python
print(data.columns)
```

```python
# Vérifier les valeurs de 'Status' dans les données originales
print(data['Status'].value_counts())
```

```python
# Convertir 'Status' en binaire
data['Status'] = data['Status'].apply(lambda x: 1 if x == 'Dead' else 0)

# Vérifier les valeurs de 'Status' après conversion
print(data['Status'].value_counts())
```

```python

# Variables catégorielles et numériques
categorical_vars = ['Race', 'Marital Status', 'T Stage ',  'N Stage', '6th Stage', 'differentiate', 'Grade', 'A Stage', 'Estrogen Status', 'Progesterone Status']
numeric_vars = ['Age', 'Tumor Size', 'Regional Node Examined', 'Reginol Node Positive', 'Survival Months']

# Visualisation univariée pour les variables catégorielle
for var in categorical_vars:
    plt.figure(figsize=(8, 6))  # Taille du graphique
    sns.countplot(data=data, x=var, hue=var, palette='viridis', legend=False)
    plt.title(f"Distribution de {var}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()  # Afficher le graphique
    plt.close() 


for var in numeric_vars:
    plt.figure(figsize=(8, 6))  # Taille du graphique
    sns.histplot(data=data, x=var, kde=True, bins=20, color='blue')
    plt.title(f"Distribution de {var}")
    plt.tight_layout()
    plt.savefig(f"univariate_numeric_{var}.png")  # Sauvegarder le graphique
    plt.show()  # Afficher le graphique
    plt.close()  
# Visualisation bivariée pour les variables catégorielles

```

```python
for var in categorical_vars:
    plt.figure(figsize=(8, 6))  # Taille du graphique
    sns.countplot(data=data, x=var, hue='Status', palette='Set2')
    plt.title(f"{var} vs Status")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(f"bivariate_categorical_{var}.png")  # Sauvegarder le graphique
    plt.show()  # Afficher le graphique
    plt.close() 

# Visualisation bivariée pour les variables numériques
for var in numeric_vars:
    plt.figure(figsize=(8, 6))  # Taille du graphique
    sns.boxplot(data=data, x='Status', y=var, hue='Status', palette='Set3', legend=False)
    plt.title(f"{var} vs Status")
    plt.tight_layout()
    plt.savefig(f"bivariate_numeric_{var}.png")  # Sauvegarder le graphique
    plt.show()  # Afficher le graphique
    plt.close() 
```

```python
# Statistiques descriptives pour les variables catégorielles
desc_categorical = {}
for var in categorical_vars:
    desc_categorical[var] = data[var].value_counts(normalize=True) * 100  # Pourcentages

# Statistiques descriptives pour les variables numériques
desc_numeric = data[numeric_vars].describe()

# Affichage
print("Statistiques descriptives pour les variables catégorielles :")
print(desc_categorical)

print("\nStatistiques descriptives pour les variables numériques :")
print(desc_numeric)
```

```python


# Initialiser le modèle
kmf = KaplanMeierFitter()

# Ajuster le modèle
kmf.fit(data['Survival Months'], event_observed=data['Status'])

# Tracer la courbe de survie
plt.figure(figsize=(8, 6))
kmf.plot_survival_function()
plt.title("Courbe de Kaplan-Meier (Survie globale)")
plt.xlabel("Temps (mois)")
plt.ylabel("Probabilité de survie")
plt.savefig("kaplan_meier_global.png")  # Sauvegarder le graphique
plt.show()
plt.close()
```

```python
# Comparer les courbes de survie par groupe (exemple : T Stage)
plt.figure(figsize=(8, 6))
for stage in data['T Stage '].unique():
    mask = data['T Stage '] == stage
    kmf.fit(data[mask]['Survival Months'], event_observed=data[mask]['Status'], label=stage)
    kmf.plot_survival_function()

plt.title("Courbes de Kaplan-Meier par T Stage")
plt.xlabel("Temps (mois)")
plt.ylabel("Probabilité de survie")
plt.legend()
plt.savefig("kaplan_meier_t_stage.png")
plt.show()
# Sauvegarder le graphique
plt.close()

from lifelines.statistics import multivariate_logrank_test

# Appliquer le test du log-rank pour plusieurs groupes
results = multivariate_logrank_test(data['Survival Months'], data['T Stage '], data['Status'])

# Afficher les résultats
print("Résultat du test du log-rank :")
print(results)
```

```python
# Tracer les courbes de survie pour chaque groupe de 'Race'
plt.figure(figsize=(10, 6))
for race in data['Race'].unique():
    mask = data['Race'] == race
    kmf.fit(data[mask]['Survival Months'], event_observed=data[mask]['Status'], label=race)
    kmf.plot_survival_function()

plt.title("Courbes de Kaplan-Meier par Race")
plt.xlabel("Temps (mois)")
plt.ylabel("Probabilité de survie")
plt.legend()
plt.show()

from lifelines.statistics import multivariate_logrank_test

# Appliquer le test du log-rank pour plusieurs groupes
results = multivariate_logrank_test(data['Survival Months'], data['Race'], data['Status'])

# Afficher les résultats
print("Résultat du test du log-rank :")
print(results)
```

```python
# Tracer les courbes de survie pour chaque groupe de 'Marital Status'
plt.figure(figsize=(10, 6))
for status in data['Marital Status'].unique():
    mask = data['Marital Status'] == status
    kmf.fit(data[mask]['Survival Months'], event_observed=data[mask]['Status'], label=status)
    kmf.plot_survival_function()

plt.title("Courbes de Kaplan-Meier par Marital Status")
plt.xlabel("Temps (mois)")
plt.ylabel("Probabilité de survie")
plt.legend()
plt.show()

# Test du log-rank pour comparer les groupes
from lifelines.statistics import multivariate_logrank_test

# Appliquer le test du log-rank pour plusieurs groupes
results = multivariate_logrank_test(data['Survival Months'], data['Marital Status'], data['Status'])

# Afficher les résultats
print("Résultat du test du log-rank :")
print(results)
```

```python
# Tracer les courbes de survie pour chaque groupe de 'Grade'
plt.figure(figsize=(10, 6))
for grade in data['Grade'].unique():
    mask = data['Grade'] == grade
    kmf.fit(data[mask]['Survival Months'], event_observed=data[mask]['Status'], label=grade)
    kmf.plot_survival_function()

plt.title("Courbes de Kaplan-Meier par Grade")
plt.xlabel("Temps (mois)")
plt.ylabel("Probabilité de survie")
plt.legend()
plt.show()

# Test du log-rank pour comparer les groupes
from lifelines.statistics import multivariate_logrank_test

# Appliquer le test du log-rank pour plusieurs groupes
results = multivariate_logrank_test(data['Survival Months'], data['Survival Months'], data['Status'])

# Afficher les résultats
print("Résultat du test du log-rank :")
print(results)
```

```python

# Encodage des variables catégorielles avec drop_first=True
data_encoded = pd.get_dummies(data, columns=['Race', 'Marital Status', 'T Stage ', 'N Stage', '6th Stage', 'differentiate', 'Grade', 'A Stage', 'Estrogen Status', 'Progesterone Status'], drop_first=True)

# Ajouter la colonne 'Status' à data_encoded
data_encoded['Status'] = data['Status']

# Nettoyer les noms de colonnes
data_encoded.columns = data_encoded.columns.str.strip()

# Vérifier les colonnes
print("Colonnes disponibles :", data_encoded.columns)

# Vérifier les valeurs manquantes dans 'Status'
print("Valeurs manquantes dans 'Status' :", data_encoded['Status'].isnull().sum())

# Vérifier le type de données de 'Status'
print("Type de données de 'Status' :", data_encoded['Status'].dtype)

# Convertir 'Status' en type numérique si nécessaire
if data_encoded['Status'].dtype != 'int64':
    data_encoded['Status'] = data_encoded['Status'].astype(int)

# Vérifier les variables constantes
constant_columns = data_encoded.columns[data_encoded.nunique() == 1]
print("Variables constantes :", constant_columns)

# Supprimer les variables constantes (sauf 'Status')
constant_columns = [col for col in constant_columns if col != 'Status']
data_encoded = data_encoded.drop(columns=constant_columns)

# Ajuster le modèle de Cox avec régularisation L2
cph = CoxPHFitter(penalizer=0.1, l1_ratio=0.0)
cph.fit(data_encoded, duration_col='Survival Months', event_col='Status')

# Afficher le résumé du modèle
print(cph.print_summary())

# Visualiser les rapports de risque (Hazard Ratios)
plt.figure(figsize=(10, 6))
cph.plot()
plt.title("Rapports de risque (Hazard Ratios)")
plt.show()
results_df = cph.summary
doc = Document()

# Ajouter un titre
doc.add_heading('Résultats de l\'analyse de survie', 0)

# Ajouter le résumé du modèle
doc.add_heading('Résumé du modèle', level=1)

# Ajouter le tableau des résultats
doc.add_heading('Tableau des résultats', level=1)
table = doc.add_table(results_df.shape[0]+1, results_df.shape[1])

# Ajouter les en-têtes
for j, col in enumerate(results_df.columns):
    table.cell(0, j).text = col

# Ajouter les données
for i, row in enumerate(results_df.itertuples(), start=1):
    for j, value in enumerate(row[1:], start=0):
        table.cell(i, j).text = str(value)

# Sauvegarder le document
doc.save('resultats_analyse_survie.docx')
```
